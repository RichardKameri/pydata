```python
# using median to deal with missing values
columns = ['Life expectancy ', 'Adult Mortality', 'Alcohol', 'Hepatitis B', ' BMI ', 'Polio', 'Diphtheria ', 'GDP', 'Population', ' thinness  1-19 years', ' thinness 5-9 years','Income composition of resources']
for col in columns:
    df[col].fillna(df[col].median(), inplace=True)
```

```python
#Using median to deal with missing values
columns = ['Total expenditure', 'Schooling' ]

for column in columns:
  df[column]=df[column].fillna(df[column].mean())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Set the style for Seaborn
sns.set(style="whitegrid")
```

Line plot of life expectancy over time for developed and developing countries

```python
# Line plot of life expectancy over time for developed and developing countries
plt.figure(figsize=(12,6))
sns.lineplot(data=df,x='Year',y='Life expectancy ',hue='Status')
plt.title('Life Expectancy Over Time for Developed and Developing Countries')
plt.xlabel('Year')
```

BOX PLOT FOR LIFE EXPECTANCY


```python
plt.figure(figsize=(12,6))
sns.boxplot(data=df,x='Status',y='Life expectancy ')
plt.title('Box Plot of Life Expectancy for Developed and Developing Countries')
plt.xlabel('Status')
```

Scatter plot of GDP vs Life Expectancy for developed and developing countries

```python
plt.figure(figsize=(12,6))
sns.scatterplot(data=df,x='GDP',y='Life expectancy ',hue='Status')
plt.title('Scatter Plot of GDP vs Life Expectancy for Developed and Developing Countries')
plt.xlabel('GDP')
```

FACTORS AFFECTING LIFE EXPECTANCY


```python
plt.figure(figsize=(12,6))
sns.barplot(data=df,x='Status',y='Life expectancy ')
plt.title('Average Life Expectancy by Country Status')
plt.xlabel('Country Status')
```

```python
# line plot of life expectancy based on gdp.schooling and immunisation

plt.figure(figsize=(12,6))
sns.lineplot(data=df,x='Year',y='Life expectancy ',hue='Status')
plt.title('Life Expectancy Over Time for Developed and Developing Countries')
plt.xlabel('Year')
```

ased on the data and visualizations, we can summarize the factors affecting life expectancy:

GDP: Higher GDP is typically associated with higher life expectancy. This trend is more pronounced in developed countries due to better healthcare infrastructure and economic stability.

Schooling: Higher education levels correlate with higher life expectancy, reflecting better health awareness and access to healthcare services. The impact is significant in both developed and developing countries.

Immunization: Higher immunization rates are crucial in developing countries, where preventable diseases significantly impact life expectancy. Developed countries generally have higher and more consistent immunization rates.

Conclusion
By contrasting these factors in developed and developing countries over time, we can identify that economic factors like GDP are significant in both, while improvements in health infrastructure (such as immunization) have a more immediate impact in developing countries.
